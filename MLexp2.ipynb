```python
import pandas as pd
df= pd.read_csv("FuelConsumption (2).csv")
print(df.head())
print(df.tail())
print("\n Column name:")
print(df.columns.to_list())
print("\n shape:")
print(df.shape)
print("\n datatype:")
print(df.dtypes)
print("\n stat:")
print(df.info())
print(df.describe(include="all"))
print("\n missing values in each column:")
print(df.isnull())
#forward fill
df_ffill=df.copy()
df_ffill['FUELCONSUMPTION_CITY']=df_ffill['FUELCONSUMPTION_CITY'].fillna(method='ffill')
print("\n missing values after forward fill:")
print(df_ffill.isnull())
#backward fill
df_bfill=df.copy()
df_bfill['FUELCONSUMPTION_CITY']=df_bfill['FUELCONSUMPTION_CITY'].fillna(method='bfill')
print("\n missing values after backward fill:")
print(df_bfill.isnull())

import pandas as pd
from sklearn.preprocessing import LabelEncoder
df_encoded = df.copy()
le=LabelEncoder()
df_encoded ['VEHICLECLASS'] = le.fit_transform(df_encoded ['VEHICLECLASS'])
print(df_encoded [['VEHICLECLASS']])
print(df_encoded [['VEHICLECLASS']].drop_duplicates())

```
